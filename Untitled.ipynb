```python
def stock_command_naver():
    
    stock_start=time.time()
    
    codes=['011070', '108320', '005930', '000100', '035720']
    my_list = ['LG이노텍', 'LX세미콘', '삼성전자', '유한양행', '카카오']
    my_list_nick = ['LT', 'LX', 'SS', 'YY', 'KA']
    price_list=[]
    percent_list=[]

    for code in codes:
        url=f'https://finance.naver.com/item/sise.naver?code={code}'
        response=requests.get(url)
        html=response.text
        soup=BeautifulSoup(html,'html.parser')
        price=soup.select_one('#_nowVal').text
        price=price.replace(',','')
        percent=soup.select_one('#_rate')
        percent=percent.text.replace('\n','').replace('\t','')
        price_list.append(price)
        percent_list.append(percent)

    my_df=pd.DataFrame(my_list,index=codes,columns=['nick'])
    my_df['종가']=price_list
    my_df['등락률']=percent_list

    print(f'{my_df[["nick","종가", "등락률"]].values.tolist()}')
    print(f'time : {time.time()-stock_start : .2f}')
    txt1.insert('1.0', f'time : {time.time()-stock_start : .2f}\n')
    
```

```python
%%time
stock_command_naver()
```
