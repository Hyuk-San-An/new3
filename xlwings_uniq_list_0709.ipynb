```python
%%time
# 7/9xlwings
# only list
import time
start1=time.time()

from openpyxl import load_workbook
from openpyxl.styles import Font, Border,Side,PatternFill,Alignment
from openpyxl.utils.cell import coordinate_from_string
import time
import win32com.client
import pandas as pd
import xlwings as xw
import excel
import time

file_name="9628_new"
wb=xw.Book(file_name+".xlsx")
result_book=xw.Book("Format_new.xlsx")
ws=wb.sheets[0]

print("Loading time :", time.time() - start1) 
start2=time.time()

def last_row_src(sh):
    last_row=0
    first_row=15
    max_row=sh.cells(10000,15).end('up').row
    for i in range(max_row):
        if str(sh.cells(first_row+i,15).value)+str(sh.cells(first_row+i,16).value)=='NoneNone':
            last_row=i+14
            break
    return last_row

data_frame_list=[]
data_frame_list_uniq=[]
df_with_qty_list=[]

sheet_list=wb.sheets
sh_name_list=[name.name for name in wb.sheets ]

Total_data_list=[[],[],[],[],[],[]]     #Total_data_list[0] : Part Name , Total_data_list[1] : Unit Price
Total_QTY_Default_list=[]
for sh_index,i in enumerate(sheet_list): 
    last_row=last_row_src(i)
 
    if i.cells(15,2).value==1 and i.cells(11,2).value=='순' : 
        Total_data_list[0].append(i.range((15,15),(last_row,15)).value)
        Total_data_list[1].append(i.range((15,16),(last_row,16)).value)
        Total_data_list[2].append(i.range((15,22),(last_row,22)).value)
        Total_data_list[3].append(i.range((15,32),(last_row,32)).value)
        Total_data_list[4].append(i.range((15,35),(last_row,35)).value)
        Total_data_list[5].append(i.range((15,37),(last_row,37)).value)
         
Total_data_series=[[],[],[],[],[],[]] 
for tot_index,tot_data  in enumerate(Total_data_list):
    for sh_index,sh_data in enumerate(tot_data):
        for ele_index, ele in enumerate(sh_data):
            Total_data_series[0].append ( Total_data_list[0][sh_index][ele_index] )
            Total_data_series[1].append ( Total_data_list[1][sh_index][ele_index] )
            Total_data_series[2].append ( Total_data_list[2][sh_index][ele_index] )
            Total_data_series[3].append ( Total_data_list[3][sh_index][ele_index] )
            Total_data_series[4].append ( Total_data_list[4][sh_index][ele_index] )
            Total_data_series[5].append ( Total_data_list[5][sh_index][ele_index] )
                                         
PN_Dup_index_list=[ index for index,i in enumerate(Total_data_series[0]) if i in Total_data_series[0][:index] ]
for ind ,dup in enumerate(PN_Dup_index_list):           
    
    for index, data in enumerate(Total_data_series):
        del data[dup-ind]     
############################################ QTY_list 생성        
QTY_list=[]           
for sh_index,i in enumerate(sheet_list):      #sheetlist 개의 원본 sheet
    QTY_sh_list=[]
    QTY_pn_list=[]
    QTY_qty_list=[]
    for data_index, data in enumerate(Total_data_series[0]):
        QTY_pn_list.append(data)
        QTY_qty_list.append(None)     
#         QTY_sh_list.append(QTY_data_list)      
    QTY_sh_list=[QTY_pn_list]+[QTY_qty_list]
    QTY_list.append(QTY_sh_list)  

############################################ QTY_list 값변경 
for sh_index,i in enumerate(sheet_list): 
    for data_index, data in enumerate(Total_data_list[0][sh_index]):
         for qty_index,qty in enumerate(QTY_list[sh_index][0]):
                if qty==data:
                    QTY_list[sh_index][1][qty_index] =Total_data_list[2][sh_index][data_index]
############################################  
    
for sh_index,i in enumerate(sheet_list): 
#         sh=result_book.copy_worksheet(result_book.worksheets[0])
#         sh.title=i.title+"_uniq"
        sh=result_book.sheets[0].copy(name=sh_name_list[sh_index]+"_uniq")
        
        sh.cells(15,15).options(transpose=True).value=Total_data_series[0]
        sh.cells(15,16).options(transpose=True).value=Total_data_series[1]
#         excel.write(sh,Total_data_series[2],(15,22))
        sh.cells(15,32).options(transpose=True).value=Total_data_series[3]
        sh.cells(15,35).options(transpose=True).value=Total_data_series[4]
        sh.cells(15,37).options(transpose=True).value=Total_data_series[5]
        sh.cells(15,22).options(transpose=True).value=QTY_list[sh_index][1]     

print("Processing time :", time.time() - start2) 
start3=time.time()

result_book.save("Format_new1.xlsx")   

print("Saving time :", time.time() - start3) 
print("Total time :", time.time() - start1) 
```
